## 02.04 ¿Qué equipo convirtio más goles en 1991?

```python
equipos = pd.unique(data_1991['equipo'])

total_goles_mayor = 0
equipo_buscado = "DEFAULT"

for equipo in equipos:
  print(equipo)
  total_goles = data_1991[data_1991['equipo'] == equipo]['goles'].sum()
  print(total_goles)
  # comparamos
  if(total_goles > total_goles_mayor):
    total_goles_mayor = total_goles
    equipo_buscado    = equipo

print("----------")
print(total_goles_mayor)
print(equipo_buscado)
```

# 03 Creamos una funcion  **equipo_total_goles**

```python
def equipo_total_goles(data):
  '''Retorna que equipo tiene más goles: su nombre y cantidad'''
  total_goles_mayor = 0
  equipo_buscado = "DEFAULT"
  for equipo in equipos:
    #print(equipo)
    total_goles = data[data['equipo'] == equipo]['goles'].sum()
    #print(total_goles)

    if(total_goles >= total_goles_mayor):
      total_goles_mayor = total_goles
      equipo_buscado    = equipo

  #print("----------")
  #print(total_goles_mayor)
  #print(equipo_buscado)

  return([equipo_buscado,total_goles_mayor])
```

## 03 .01 Testeamos con data_1991

```python
equipo_total_goles(data_1991)
```

#04 Utilizarlo para todo el dataset

```python
for anio in anios:
  data = resultados_cmff[resultados_cmff['anio'] == anio]
  resultado = equipo_total_goles(data)
  print([anio,resultado])
```

# *Opción 2*: Utilizando Pandas groupby & .iloc

# 05 Groupby Pandas

## 05.01 sumamos goles por anio y equipo

```python
resultados_cmff.groupby(['anio','equipo'])['goles'].sum()
```

## 05.02 Creamos un nuevo dataframe

```python
 df_goles = resultados_cmff.groupby(['anio','equipo'])['goles'].sum().reset_index(name ='total_goles')
 df_goles
```

## 05.03 ¿Cuáles son los ***valores máximos***  de **total_goles** ?

```python
df_goles.groupby(['anio'])['total_goles'].max()
```

## 05.04 Averiguemos en que **posiciones** del dataframe  se encuentran esos valores **máximos**

```python
 id_maximo = df_goles.groupby(['anio'])['total_goles'].idxmax()
```

## 05.05 Encontremos los equipos según el id_ de posición, utilizando **.iloc**

```python
df_goles.iloc[id_maximo]
```
